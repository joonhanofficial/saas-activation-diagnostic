```python
accounts["signup_date"] = pd.to_datetime(accounts["signup_date"])
subscriptions["start_date"] = pd.to_datetime(subscriptions["start_date"])
feature_usage["usage_date"] = pd.to_datetime(feature_usage["usage_date"])
```

```python
print("accounts signup_date type:", accounts["signup_date"].dtype)
print("subscriptions start_date type:", subscriptions["start_date"].dtype)
print("feature_usage usage_date type:", feature_usage["usage_date"].dtype)
```

```python
first_usage = (
    feature_usage
    .groupby("subscription_id", as_index=False)["usage_date"]
    .min()
    .rename(columns={"usage_date": "first_usage_date"})
)
```

```python
first_usage.head(10)
```

```python
subscriptions = subscriptions.merge(first_usage, on="subscription_id", how="left")
```

```python
subscriptions[["subscription_id", "start_date", "first_usage_date"]].head(10)
```

```python
invalid_usage_timing = subscriptions["first_usage_date"] < subscriptions["start_date"]
print("Subscriptions with first usage earlier than start date:", invalid_usage_timing.sum())
```

```python
usage_summary = (
    feature_usage
    .groupby("subscription_id", as_index=False)
    .agg(
        total_usage_count=("usage_count", "sum"),
        total_usage_duration_secs=("usage_duration_secs", "sum"),
        total_error_count=("error_count", "sum"),
        distinct_features_used=("feature_name", "nunique"),
        usage_records=("usage_id", "count")
    )
)
```

```python
usage_summary.head(10)
```

```python
subscriptions = subscriptions.merge(usage_summary, on="subscription_id", how="left")
```

```python
subscriptions[[
    "subscription_id",
    "account_id",
    "plan_tier",
    "is_trial",
    "mrr_amount",
    "total_usage_count",
    "total_usage_duration_secs",
    "distinct_features_used",
    "usage_records"
]].head(10)
```

```python
subscriptions[[
    "total_usage_count",
    "total_usage_duration_secs",
    "total_error_count",
    "distinct_features_used",
    "usage_records"
]] = subscriptions[[
    "total_usage_count",
    "total_usage_duration_secs",
    "total_error_count",
    "distinct_features_used",
    "usage_records"
]].fillna(0)
```

```python
subscriptions["activated_flag"] = (
    (subscriptions["usage_records"] >= 3) &
    (subscriptions["distinct_features_used"] >= 3)
)
```

```python
subscriptions["activated_flag_v2"] = (
    (subscriptions["usage_records"] >= 5) &
    (subscriptions["distinct_features_used"] >= 5)
)
```

```python
subscriptions["activated_flag"] = subscriptions["activated_flag_v2"]
```

```python
subscriptions.groupby("plan_tier")["activated_flag"].mean().sort_values(ascending=False)
```

```python
subscriptions = subscriptions.merge(
    accounts[["account_id", "referral_source"]],
    on="account_id",
    how="left"
)

subscriptions.groupby("referral_source")["activated_flag"].mean().sort_values(ascending=False)
```

```python
subscriptions = subscriptions.merge(
    accounts[["account_id", "industry"]],
    on="account_id",
    how="left"
)

subscriptions.groupby("industry")["activated_flag"].mean().sort_values(ascending=False)
```

```python
subscriptions.groupby("referral_source").agg(
    total_subscriptions=("subscription_id", "count"),
    activated_subscriptions=("activated_flag", "sum"),
    activation_rate=("activated_flag", "mean")
).sort_values("activation_rate", ascending=False)
```

```python
subscriptions.groupby("plan_tier").agg(
    total_subscriptions=("subscription_id", "count"),
    activated_subscriptions=("activated_flag", "sum"),
    activation_rate=("activated_flag", "mean")
).sort_values("activation_rate", ascending=False)
```

```python
subscriptions.pivot_table(
    index="referral_source",
    columns="plan_tier",
    values="activated_flag",
    aggfunc="mean"
)
```
